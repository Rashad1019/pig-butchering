```python
import matplotlib.pyplot as plt
from pathlib import Path

# ── Define YOUR exact output directory ──
output_dir = Path(r"C:\Users\rasha_ejuf17z\Downloads\0Today\1. Business\1. Data\pig_butchering\output")
output_dir.mkdir(parents=True, exist_ok=True)

# ── Headline metrics ──
annual_loss     = "44 B"      # USD-billions (2024)
forced_workers  = "313 k"     # people
crypto_conv     = "93 %"      # %
proj_loss       = "142.8 B"   # USD-billions
regions = [
    ("SE Asia",       "21.0 B"),
    ("United States", "9.4 B"),
    ("Australia",     "4.7 B")
]

# ── Build figure ──
fig = plt.figure(figsize=(9, 5))
fig.patch.set_facecolor('white')
ax = fig.add_subplot(111)
ax.axis('off')

# ── Positions for the four indicators ──
positions = [
    (0.20, 0.80, "Annual Losses\n(2024)",      annual_loss),
    (0.70, 0.80, "Trafficked Workers",         forced_workers),
    (0.20, 0.40, "Crypto Converted ≤ 48 h",    crypto_conv),
    (0.70, 0.40, "Projected Losses\n(2025)",  proj_loss),
]

# ── Draw metrics ──
for x, y, title, value in positions:
    fig.text(x, y + 0.10, title,
             ha='center', va='center', fontsize=12, weight='bold')
    fig.text(x, y - 0.05, value,
             ha='center', va='center', fontsize=32)

# ── Bottom list of regions ──
region_text = "Top Impacted Regions\n" + "\n".join(f"{r}: {v}" for r, v in regions)
fig.text(0.50, 0.05, region_text,
         ha='center', va='center', fontsize=10)

plt.tight_layout()

# ── Overwrite the old Plotly PNG directly ──
out_path = output_dir / "executive_summary_panel.png"
fig.savefig(str(out_path), dpi=300, bbox_inches='tight')

print(f"✅ Executive summary panel written to:\n  {out_path}")
```
